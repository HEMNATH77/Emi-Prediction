```python
import joblib
import os

# --- File paths (your existing model files) ---
REG_MODEL_PATH = "Final_Regression.pkl"
CLS_MODEL_PATH = "FinalClassification.pkl"

# --- Compressed output file names ---
REG_MODEL_COMPRESSED = "Final_Regression_Cp.pkl"
CLS_MODEL_COMPRESSED = "Final_Classification_Cp.pkl"

# --- Compress and save with gzip (level 3 = good balance between speed & size) ---
print("Compressing Regression model...")
reg_model = joblib.load(REG_MODEL_PATH)
joblib.dump(reg_model, REG_MODEL_COMPRESSED, compress=('lzma', 3))
print("✅ Regression model compressed and saved as:", REG_MODEL_COMPRESSED)

print("Compressing Classification model...")
cls_model = joblib.load(CLS_MODEL_PATH)
joblib.dump(cls_model, CLS_MODEL_COMPRESSED, compress=('lzma', 3))
print("✅ Classification model compressed and saved as:", CLS_MODEL_COMPRESSED)

# --- Print file size comparison ---
def size_in_mb(path):
    return round(os.path.getsize(path) / (1024 * 1024), 2)

print("\n--- File size comparison (MB) ---")
print(f"Regression (Original): {size_in_mb(REG_MODEL_PATH)} MB → Compressed: {size_in_mb(REG_MODEL_COMPRESSED)} MB")
print(f"Classification (Original): {size_in_mb(CLS_MODEL_PATH)} MB → Co mpressed: {size_in_mb(CLS_MODEL_COMPRESSED)} MB")
```
